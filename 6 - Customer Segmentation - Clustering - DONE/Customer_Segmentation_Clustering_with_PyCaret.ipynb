```python
import pandas as pd
from pycaret.clustering import *
```

```python
df=pd.read_csv("PBL5recommendationdata.csv", encoding = "ISO-8859-1")
```

```python
setup(df) # TypeError: '<' not supported between instances of 'int' and 'str'
```

'<' not supported between instances of 'int' and 'str': böyle bir hata veriyor. ondan dolayi df_new ile dene

```python
#remove all object datatypes and assign it to a new dataframe
df_new=df.select_dtypes(exclude="object") 
```

```python
#Remove missing values.
#‘all’ : If all values are NA, drop that column
df_new.dropna(axis=1,how="all",inplace=True)
```

```python
for i in df_new:
    df_new[i].fillna(df_new[i].mean(),inplace=True)
```

```python
setup(df_new)
```

```python
kmeans=create_model('kmeans',num_clusters=4)
```

```python
plot_model(kmeans)
```

```python
plot_model(kmeans, feature='Orders.total', label=True)
```

```python
plot_model(kmeans,plot='elbow')
```

```python
plot_model(kmeans,plot='tsne')
```
