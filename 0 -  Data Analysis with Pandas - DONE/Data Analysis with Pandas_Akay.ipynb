### <font color=blue> Complex Queries¶

```python
#clients older than 30 and female
df[(df.AGE>30) & (df.GENDER=='F')]
```

```python
#clients older than 30 and female but only hair and eye color
df[(df.AGE>30) & (df.GENDER=='F')][['HAIR_COLOR','EYE_COLOR']]
```

```python
#unique combination of hair and eye color for women older than 25 years
df[(df.GENDER=='F') & (df.AGE>25)][['HAIR_COLOR','EYE_COLOR']].drop_duplicates()
```

### <font color=blue>Additional Dataframe Operations

```python
#merge two files
df1=pd.read_csv("client_list.csv")
df2=pd.read_csv("customer_id_list.csv")
clients=pd.concat([df1,df2])
clients
```

```python
#merge two files only who pucharsed
df3=pd.read_csv("purchase_log.csv")
detailed_sales=pd.concat([clients,df3])
detailed_sales[['CUSTOMER_ID']]
```

```python
#find most spender with groupby()
spenders = detailed_sales.groupby(['CUSTOMER_ID'], as_index=False)[['PRICE']].sum()
max_spender = spenders[spenders.PRICE==spenders.PRICE.max()]
max_spender
```

```python
#find item of most spender
item = max_spender.reset_index(drop=True).loc[0,'CUSTOMER_ID']
df3.loc[df3.CUSTOMER_ID==item, 'ITEM_DESCRIPTION']
```

### <font color=blue>Writing Files

```python
#save csv file with no indicies
detailed_sales.to_csv("df_out.csv", index=False)
```
